# Labels Info

- 0 = Buy
- 1 = Sell
- 2 = Hold/Do Nothing

12 hour labelling script
```python
price_df.to_csv("labels.csv")
from tqdm import tqdm
for i in tqdm(range(len(price_df))):
    if(i%1000==0):
        price_df.to_csv("labels.csv")
    try:
        price_df['Price_Present'][i]=daily_price.loc[(daily_price.Ticker==fundamen_data.loc[i]['Ticker'] )& (daily_price['Date DT'] >= fundamen_data.loc[i]['Publish Date DT'])]['Close'].iloc[0]
        price_df['Price_Future'][i]=daily_price.loc[(daily_price.Ticker==fundamen_data.loc[i]['Ticker'] )& (daily_price['Date DT'] >= (fundamen_data.loc[i]['Publish Date DT']+pd.DateOffset(months=3)))]['Close'].iloc[0]
        curr=price_df['Price_Present'][i]
        fut=price_df['Price_Future'][i]
        if fut>=(curr+0.03*curr):
            price_df['Label'][i]=0
        elif fut<=(curr-0.03*curr):
            price_df['Label'][i]=1
        else:
            price_df['Label'][i]=2
    except:
        price_df['Price_Present'][i]=np.NaN
        price_df['Price_Future'][i]=np.NaN
```
