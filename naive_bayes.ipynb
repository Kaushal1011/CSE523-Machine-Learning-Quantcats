```python
# import datasets
X = pd.read_csv("dataset_normalised.csv")
y = pd.read_csv("labnf.csv")

# numberical column that are needed for training
numeric_columns=['Shares (Basic)',
       'Shares (Diluted)', 'Cash, Cash Equivalents & Short Term Investments',
       'Accounts & Notes Receivable', 'Inventories', 'Total Current Assets',
       'Property, Plant & Equipment, Net',
       'Long Term Investments & Receivables', 'Other Long Term Assets',
       'Total Noncurrent Assets', 'Total Assets', 'Payables & Accruals',
       'Short Term Debt', 'Total Current Liabilities', 'Long Term Debt',
       'Total Noncurrent Liabilities', 'Total Liabilities',
       'Share Capital & Additional Paid-In Capital', 'Treasury Stock',
       'Retained Earnings', 'Total Equity', 'Total Liabilities & Equity',
       'Price_Present']
X = X[numeric_columns]
X.head()
```

## 1.Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()

kf = KFold(n_splits=5,shuffle=True)

scores = []
cm_holder = []

for train_index,test_index in kf.split(X):
    X_train,X_test = X.iloc[train_index],X.iloc[test_index]
    y_train,y_test = y.iloc[train_index],y.iloc[test_index]
        
    clf.fit(X_train,y_train)
    
    train_score = clf.score(X_train,y_train)
    test_score = clf.score(X_test,y_test)
    
    scores.append((train_score,test_score))
    
    cm_holder.append(confusion_matrix(y_test,clf.predict(X_test)))
    
print("Accuracy :\tTrain\t\t\tTest\n")
train_avg = 0
test_avg = 0
for i in range(len(scores)):
    train_avg += scores[i][0]
    test_avg += scores[i][1]
    print("Split {}:\t{}%\t{}%".format(i,scores[i][0]*100,scores[i][1]*100))

print("\nMean :\t\t{}%\t{}%".format(train_avg/len(scores)*100,test_avg/len(scores)*100))

final_cm = np.zeros((3,3))
for cm in cm_holder:
    final_cm += cm

# print averaged confusion matrix

print("\nConfusion Matrix:\n\n",final_cm/len(cm_holder))


```
